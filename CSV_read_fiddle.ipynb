```python
import pandas as pd
from nltk.tokenize import word_tokenize

data = pd.read_csv('rianru_tweets.csv', delimiter='|', header=None)
```

```python
dic = {}
for line in data.values:
    try:
        tweet = word_tokenize(line[3].decode('utf-8'))
    except:
        tweet = []
        print("Error tokenizing %s: encoding issues" % line[3])
    for token in tweet:
        if token in dic.keys():
            dic[token] += 1
        else:
            dic[token] = 1
```

```python
import operator
for k, v in sorted(dic.items(), key=operator.itemgetter(1)):
    print("%s: %d" % (k, v))
```
